## Import Libraries

```python
import pandas as pd
from sklearn.model_selection import *
from sklearn.metrics import *
import xgboost as xgb
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import accuracy_score
```

## I used Kaggle csv dataset, Train-Test split is already there

```python
df = pd.read_csv("mnist_train.csv")
x_train = df.drop(["label"], axis = 1)
y_train = df["label"]
```

```python
df = pd.read_csv("mnist_test.csv")
x_test = df.drop(["label"], axis = 1)
y_test = df["label"]
```

## Define XGBoost model and parameter grid for hyperparameter tuning

```python

xgb_model = xgb.XGBClassifier(random_state=42)
params_grid = {'n_estimators': [50, 100, 150], 'learning_rate': [0.1, 0.2, 0.3], 'max_depth': [3, 4, 5]}

xgb_model = xgb.XGBClassifier(
    random_state=42,
    tree_method='gpu_hist',
    n_gpus=-1 # Use all available GPUs
)

params_grid = {'n_estimators': [50, 100, 150], 'learning_rate': [0.1, 0.2, 0.3], 'max_depth': [3, 4, 5]}
```

## Define GridSearchCV object and fit on training data

```python

gs_cv = GridSearchCV(xgb_model, params_grid, cv=5)
gs_cv.fit(x_train, y_train)


# Print the best hyperparameters and corresponding mean test score
print('Best hyperparameters:', gs_cv.best_params_)
print('Best mean test score: {:.2f}%'.format(gs_cv.best_score_*100))
```

## Predict on test data using the best model

```python
best_model = gs_cv.best_estimator_
y_pred = best_model.predict(x_test)
```

## Calculate accuracy score

```python
accuracy = accuracy_score(y_test, y_pred)
print('Accuracy: {:.2f}%'.format(accuracy*100))
```

## Cross Val Score score

```python
# perform cross-validation
from sklearn.model_selection import cross_val_score
cv_scores = cross_val_score(best_model, x_test, y_test, cv=5)

def Average(avg):
    return sum(avg) / len(avg)

average = Average(cv_scores)
print(average)
```

## Exporting the pickle file

```python
import pickle
filename = 'xgb_model.pickle'

# Dump the model to a pickle file
with open(filename, 'wb') as f:
    pickle.dump(best_model, f)
```
